```python
from scipy.optimize import minimize

def transform_points(points, dx, dy, dtheta):
    # Apply rotation and translation
    rotation_matrix = np.array([[np.cos(dtheta), -np.sin(dtheta)], 
                                [np.sin(dtheta),  np.cos(dtheta)]])
    return np.dot(points, rotation_matrix.T) + np.array([dx, dy])

def least_squares_adjustment(original_points, new_points):
    def error_function(params):
        dx, dy, dtheta = params
        transformed_points = transform_points(original_points, dx, dy, dtheta)
        return np.sum((transformed_points - new_points) ** 2)

    initial_guess = [0, 0, 0]  # Initial guess for dx, dy, and dtheta
    result = minimize(error_function, initial_guess, method='L-BFGS-B')
    return result.x
```

```python
original_points = df[df.pt_number.isin(cdf.pt_number)][["easting","northing"]].values
```

```python
new_points = cdf[["easting","northing"]].values
```

```python
new_points
```

```python
original_points
```

```python
dx, dy, dtheta = least_squares_adjustment(original_points, new_points)
```

```python
df[['easting', 'northing']] = transform_points(df[['easting', 'northing']].values, dx, dy, dtheta)
```
